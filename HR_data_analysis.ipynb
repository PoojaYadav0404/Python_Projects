# identify departments with the highest attrition rate(resigned %)

```python
dept_count = df.groupby("Department")["Status"].agg(total_emp = "count", resigned = lambda x: (x == "Resigned").sum())
dept_count
```

```python
dept
```

```python
type(dept_count)
```

```python
#calculating attrition rate

dept_count["Attrition %"] = (dept_count["resigned"] / dept_count["total_emp"]) * 100
```

```python
dept_count
```

```python
dept_count.sort_values("Attrition %", ascending = False)
```
