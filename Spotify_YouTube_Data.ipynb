# questions

```python
# top 10 artists with the highest views on youtube?
df.groupby("Artist")["Views"].sum().sort_values(ascending = False).reset_index().head(10)
```

```python
# top 10 tracks with the highest streams on spotify? 
x = df[["Track", "Stream"]]
```

```python
# what are the most common album types on spotify? how many tracks belong to each album type
df.groupby("Album_type")["Track"].count().sort_values(ascending = False)
```

```python
df["Album_type"].value_counts().plot(kind = "pie", autopct = "%.1f%%", explode = [.1, .1, .1], shadow = True, startangle = 60)
plt.show()
```

```python
# how do the average views , likes , & comments are compared between different album types?
y = df.groupby("Album_type")[["Views", "Likes", "Comments"]].mean()
y
```

```python
unpivot_y = y.reset_index()
unpivot_y
```

```python
mlt = pd.melt(unpivot_y, id_vars = "Album_type", var_name = "Attribte", value_name = "Total")
mlt
```

```python
plt.figure(figsize = (10, 6))
ax = sns.barplot( x = "Album_type", y = "Total" , hue = "Attribte", data = mlt)
plt.grid()
plt.show()
```

```python
# top 5 youtube channels based on the views?
y5 = df.groupby("Channel")["Views"].sum().sort_values(ascending = False).head(5).apply(int).reset_index()
y5
```

```python
yx = sns.barplot( x = "Views", y = "Channel", data = y5) # hue = "Channel")
plt.grid()
plt.bar_label(yx.containers[0])
plt.show()
```

```python
# the top most track based on views? # can apply sort and groupby both depending upon situation
t_v = df[["Track", "Views"]]
mtv = t_v.sort_values(by = "Views", ascending = False).head(3)
mtv
```
