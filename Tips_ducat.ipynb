```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#matplotlib.pylot- to visualise

tips = sns.load_dataset("tips")
tips
```

```python
#other dataset of seaborn - tips, iris, titanic, diamonds
```

```python
#line plot

x = np.linspace(0,20, 100)
y = np.exp(x)
print(x)
print("value of y" , y)

```

```python
plt.plot(x, y)
plt.show()
```

```python
np.linspace(1, 10, 4)
```

```python
x = np.linspace(0, 20, 100)
y = np.exp(x)
plt.plot(x, y, "r")
plt.show()
```

```python
x = np.linspace(0, 20, 100)
y_sqrt = np.sqrt(x)
y_sqr = np.square(x)
plt.plot(x, y_sqrt, "r", x, y_sqr, "g")
plt.show()
```

```python
#scatter plot

x = tips["total_bill"]
y = tips["tip"]
plt.figure(figsize=(10, 5))
plt.scatter(x, y, s = np.array(y)**2, color = "g")
plt.ylabel("Tips")
plt.xlabel("Total Bill")
plt.show()
```

```python
#box plot

plt.boxplot(x = "total_bill", data=tips)
plt.show()
```

```python
sns.boxplot( x = "day", y = "tip", data = tips)
plt.show()
```

```python
tips["smoker"].unique()
```

```python
#bar plot

tips["smoker"].value_counts().plot(kind = "bar")
plt.show()
```

```python
tips["day"].unique()
```

```python
#pieplot

tips["day"].value_counts().plot(kind = "pie", autopct = "%.1f%%", labels = ['Sun', 'Sat', 'Thur', 'Fri'], explode = [0.1,0, 0,0])
plt.show()
```

```python
#histogram

plt.hist("tip", data = tips, bins=5)
plt.ylabel('tips')
plt.show()
```

```python
#same argumentation(x, y, data)- swarmplot, box plot, violin plot, stripplot, regplot
```

```python
#swarm plot

sns.swarmplot(x= "smoker", y = "tip", data = tips, hue="sex")
plt.show()
```

```python
sns.violinplot(x = "day", y = "total_bill", data = tips, hue= "sex")
plt.show()
```

```python
sns.violinplot(x = "day", y = "total_bill", data = tips, hue= "sex", split= True)
plt.show()
```

```python
sns.stripplot(x = "day", y = "total_bill", data = tips)
plt.show()
```

```python
tips["time"].unique()
```

```python
sns.countplot(x = "day", data = tips, hue = "time")
plt.show()
```

```python
sns.distplot(tips["total_bill"])
plt.show()
```

```python
sns.regplot(x = "total_bill", y = "tip", data = tips)
plt.show()
```
